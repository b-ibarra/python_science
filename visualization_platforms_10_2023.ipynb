Este notebook analiza datos relacionados con los empleados de una empresa y variables potencialmente relacionadas con la tasa de deserción. El análisis y visualización de la información se realizaron mediante las paqueterías Pandas, Numpy, Matplotlib y Seaborn. El notebook incluye las siguientes etapas:  

* *Instalación* de las paqueterías.  
* *Exploración* de datos.  
* *Visualización* de datos.

### Instalación

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
# Cálculo del porcentaje de valores faltantes
perc_missing_values = (employees.isna().sum().sum() / employees.size) * 100
print(f'El % de valores faltantes en el dataframes es {perc_missing_values:.1f}%.')
```

```
employees = employees.dropna()
employees.isna().sum()
```

```
# Proporción de valores únicos
educ_level = employees['Education_Level'].nunique()
hometown = employees['Hometown'].nunique()
print(f'Existen {educ_level} niveles educativos.')
print(f'Participaron {hometown} ciudades en el estudio.')
```

```
# Conteo de valores por categoría.
gender_count = employees['Gender'].value_counts()
gender_count
```

```
status = employees['Relationship_Status'].value_counts()
status
```

```
employees_stats = employees.describe().T
employees_stats
```

```
print('El tiempo máximo de servicio es:', employees_stats.loc['Time_of_service', 'max'], 'años.')
print('La edad promedio de los empleados encuestados es:', employees_stats.loc['Age', 'mean'], 'años.')
print('El valor del primer cuartil en la escala de pago:', employees_stats.loc['Pay_Scale', '25%'],'.')
```

### Visualización de datos

```
sns.histplot(x=employees['Age'], bins=[15,20,30,40,50,60,70], color='plum', edgecolor=None)
plt.show()
```

```
print('El rango de edad con la mayor cantidad de registros es de 20 a 30 años.')
```

```
unit = employees['Unit'].value_counts()
unit_counts = pd.DataFrame({'Unit': unit.index, 'Count': unit.values})
unit_counts
```

```
ax = sns.barplot(unit_counts, x='Unit', y='Count', palette='Set2')
ax.tick_params(axis='x', rotation=90)
ax.bar_label(ax.containers[0], fontsize=10)
plt.show()
```

```
print('La unidad con mayor número de empleados es IT.')
```

```
employees_by_hometown = employees.groupby('Hometown').mean()
employees_by_hometown
```

```
employees_by_hometown.plot()
```

```
employees_by_hometown.loc[:,'Age':'Attrition_rate'].plot(subplots=True, layout=(4,3), figsize=(15,10), rot=90)
plt.show()
```

```
print('La ciudad con mayor índice de deserción es Springfield.')
```

```
sns.boxplot(data=employees, x='Education_Level', y='Pay_Scale', palette='Set2')
plt.show()
```

```
print('La escala de pago respecto al nivel educativo se distribuye de forma homogénea. \n'
           'Sin embargo, a partir del nivel educativo 3, se nota un incremento en el Q1. \n'
           'Esto implica que el 25% de los empleados con niveles educativos 3, 4 y 5, '
           'se encuentran en la escala de pago 5.')
```

```
fig, axs = plt.subplots(1,2, figsize=(10,5))
sns.scatterplot(x=employees['Attrition_rate'], y=employees['Age'], ax=axs[0], color='lightsteelblue')
sns.scatterplot(x=employees['Attrition_rate'], y=employees['Time_of_service'], ax=axs[1], color='lightslategrey')
plt.show()
```

```
print('No existe una correlación entre la tasa de deserción y las variables edad y tiempo de servicio.')
```

```
sns.heatmap(round(employees.corr(numeric_only=True),2), annot=True)
```
